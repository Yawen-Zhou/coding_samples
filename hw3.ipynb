```python
import pandas as pd
import numpy as np
kirb21_df = pd.read_csv("https://raw.githubusercontent.com/smart-stats/ds4bio_book/main/book/assetts/kirby21AllLevels.csv")
kirb21_df
```

```python
kirby906a_ax_df = kirb21_df.loc[kirb21_df['rawid'] == 'kirby906a_ax.img'].copy()
kirby906a_ax_df.head(20)
```

```python
t1l1 = kirby906a_ax_df.loc[(kirby906a_ax_df['type'] == 1) & (kirby906a_ax_df['level'] == 1)].copy()
t1l1
```

```python
telencephalon_L_volume = sum(t1l1.loc[t1l1['roi'] == 'Telencephalon_L'].volume)
telencephalon_L_volume
```

```python
telencephalon_R_volume = sum(t1l1.loc[t1l1['roi'] == 'Telencephalon_R'].volume)
telencephalon_R_volume
```

```python
telecephalon = telencephalon_L_volume + telencephalon_R_volume
ICV = sum(t1l1.loc[t1l1['roi'] == 'Telencephalon_L'].icv)
ICV_fraction = telecephalon / ICV
ICV_fraction
```

```python
TBV = sum(t1l1.loc[t1l1['roi'] == 'Telencephalon_L'].tbv)
TBV_fraction = telecephalon / TBV
TBV_fraction
```

```python
gene_expression_df = pd.read_csv("https://raw.githubusercontent.com/jhu-advdatasci/2018/master/data/GSE5859_exprs.csv", index_col=0)
gene_expression_df
```

```python
gene_expression_df_pt1 = gene_expression_df.apply(lambda row: row - row.mean(), axis=1)
gene_expression_df_pt1
```

```python
gene_expression_df_pt2 = gene_expression_df_pt1.apply(lambda column: column - column.mean(), axis=0)
gene_expression_df_pt2
```

```python
gene_expression_df_pt3 = gene_expression_df_pt2.apply(lambda column: column / column.std(), axis=0)
gene_expression_df_pt3
```

```python
healthcare_df = pd.read_csv("https://raw.githubusercontent.com/jhu-advdatasci/2018/master/data/KFF/healthcare-spending.csv", skiprows = 2)
healthcare_df.head(20)
```

```python
healthcare_df_drop = healthcare_df[1:-12]
healthcare_df_long = healthcare_df_drop.melt(id_vars=['Location'], var_name='Year', value_name='Total Health Spending')
healthcare_df_long['Year'] = healthcare_df_long['Year'].str.extract('(\d+)').astype(int)
healthcare_df_long
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(x='Year', y='Total Health Spending', hue='Location', marker='o', data=healthcare_df_long)

plt.title('Healthcare Spending Over Time by State')
plt.xlabel('Year')
plt.ylabel('Total Health Spending')
plt.legend(title='Location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
healthcare_df_average = healthcare_df_long.groupby(['Location'])['Total Health Spending'].mean().reset_index().copy()
healthcare_df_average
plt.figure(figsize=(14, 10)) 
plt.barh(healthcare_df_average['Location'], healthcare_df_average['Total Health Spending'], color = 'skyblue') 
plt.xlabel('Average Total Health Spending') 
data = healthcare_df
```
